```python
# імпорт необхідних бібліотек
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# дані:
a, b, c = 2, 1, 0
x = np.arange(0, 20)
y = a * x + b + np.random.normal(0, 1 + c, 20)
x=x[:, None]
```

```python
model = LinearRegression() # модель: лінійна регресія
model.fit(x, y) # підігнати модель (навчання або пошук коефіцієнтів)
```

```python
print("a=%f b=%f"%(model.coef_[0], model.intercept_)) # коефіцієнти моделі
print("R2=%f"%model.score(x,y)) # коефіцієнт детермінації
```

```python
X = np.linspace(0, 20, 100) # нові дані X
Y = model.predict(X[:, None]) # прогноз для X
```

```python
x_ = 17 # довільна точка
y_ = model.predict([[x_]]) # прогноз в довільній точці
print(y_)
```

```python
plt.scatter(x, y) # емпіричні дані
plt.plot(X, Y) # модель
plt.scatter(x_, y_)
plt.xlabel('x')
plt.ylabel('y')
plt.show() # будуємо графік
```
